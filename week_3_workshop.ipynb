Step 3: Handling Outliers

```python
# prompt: detect any outliers in the data and make a small boxplot and small histogram to spot these outliers. printed after all charts, print a summary list or table of number of all outliers in dataset as well as the values to ask user what outliers (or all) they want to remove. document the method you chose to remove them. comment on steps as you go and provide an overview summary at the end

import matplotlib.pyplot as plt
import seaborn as sns

# Function to detect outliers using the IQR method
def detect_outliers_iqr(data):
  """
  Detects outliers in a dataset using the Interquartile Range (IQR) method.

  Args:
    data: A pandas Series or NumPy array.

  Returns:
    A list of outlier values.
  """
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data[(data < lower_bound) | (data > upper_bound)]
  return outliers.values.tolist()


# Create a dictionary to store outlier information
outlier_summary = {}

# Iterate through each numerical column in the DataFrame
for column in df.select_dtypes(include=['number']).columns:
  # Detect outliers using the IQR method
  outliers = detect_outliers_iqr(df[column])

  # Store outlier information in the dictionary
  outlier_summary[column] = {
      "count": len(outliers),
      "values": outliers
  }

  # Create a boxplot to visualize outliers
  plt.figure(figsize=(4, 3))  # Adjust figure size as needed
  sns.boxplot(x=df[column])
  plt.title(f'Boxplot of {column}')
  plt.show()

  # Create a histogram to visualize outliers
  plt.figure(figsize=(4, 3))  # Adjust figure size as needed
  sns.histplot(df[column], kde=True)
  plt.title(f'Histogram of {column}')
  plt.show()


# Print the summary of outliers
print("Summary of outliers:")
for column, info in outlier_summary.items():
  print(f"{column}:")
  print(f"  Number of outliers: {info['count']}")
  print(f"  Outlier values: {info['values']}")


# Get user input on which outliers to remove
print("\nDo you want to remove any outliers?")
print("1. Remove all outliers")
print("2. Remove specific outliers (enter column names separated by commas)")
print("3. Do not remove any outliers")
choice = input("Enter your choice (1, 2, or 3): ")

if choice == "1":
  # Remove all outliers
  for column in outlier_summary:
    if outlier_summary[column]["count"] > 0:
      # Get indices of outliers
      outlier_indices = df[df[column].isin(outlier_summary[column]["values"])].index
      # Drop rows containing outliers
      df.drop(outlier_indices, inplace=True)
  print("All outliers removed.")

elif choice == "2":
  # Remove specific outliers
  columns_to_remove = input("Enter column names separated by commas: ").split(",")
  for column in columns_to_remove:
    column = column.strip()  # Remove leading/trailing whitespace
    if column in outlier_summary and outlier_summary[column]["count"] > 0:
      # Get indices of outliers
      outlier_indices = df[df[column].isin(outlier_summary[column]["values"])].index
      # Drop rows containing outliers
      df.drop(outlier_indices, inplace=True)
  print("Specific outliers removed.")

elif choice == "3":
  print("No outliers removed.")
else:
  print("Invalid choice.")

# Overview summary
print("\nOverview Summary:")
print("Outliers detected and handled.")
print("Method used to remove outliers: IQR method.")
print("Note: Outlier removal can affect data distribution. Carefully consider the impact on your analysis.")
```

Step 9: Dealing with Multicollinearity

```python
# prompt: ID pairs of highly correlated vars (close to 1 or -1) to ID which features might distort the model. decode on if you chould drop one of the correlated variables or combine correlated variables into a new feature. explain what each choice will do to model, reccommend what to do and allow user to pick which to do, explain each step & explain each choice impact to user. allow user to undo

# Identify pairs of highly correlated variables
correlation_threshold = 0.7  # Define a threshold for high correlation

highly_correlated_pairs = []
for i in range(len(correlation_matrix.columns)):
  for j in range(i + 1, len(correlation_matrix.columns)):
    if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
      highly_correlated_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))

# Print highly correlated pairs
print("\nHighly Correlated Pairs (above threshold of", correlation_threshold, "):")
for col1, col2, correlation in highly_correlated_pairs:
  print(f"  {col1} and {col2}: {correlation:.2f}")

# Explain the potential issues with multicollinearity
print("\nMulticollinearity can cause issues in statistical models:")
print("  - Increased standard errors of coefficients.")
print("  - Difficulty in interpreting the individual effects of variables.")
print("  - Reduced model stability and prediction accuracy.")

# Explain options for dealing with multicollinearity
print("\nOptions for dealing with multicollinearity:")
print("  1. Drop one of the correlated variables.")
print("    - This simplifies the model and reduces redundancy.")
print("    - It can improve model interpretability and stability.")
print("    - However, it may lead to loss of information.")
print("  2. Combine correlated variables into a new feature.")
print("    - This retains information from both variables.")
print("    - It can improve model performance by capturing the combined effect.")
print("    - However, it can increase complexity and may require careful feature engineering.")

# Get user input on how to handle multicollinearity
print("\nHow do you want to handle multicollinearity?")
print("1. Drop one of the correlated variables")
print("2. Combine correlated variables into a new feature")
choice = input("Enter your choice (1 or 2): ")

if choice == "1":
  # Drop one of the correlated variables
  print("\nYou chose to drop one of the correlated variables.")
  print("Please select the variable to drop from each pair:")
  for col1, col2, correlation in highly_correlated_pairs:
    print(f"  - {col1} and {col2}:")
    drop_choice = input(f"    Enter the variable to drop ({col1} or {col2}): ")
    if drop_choice == col1:
      df_encoded.drop(columns=col1, inplace=True)
      print(f"    Dropped {col1}")
    elif drop_choice == col2:
      df_encoded.drop(columns=col2, inplace=True)
      print(f"    Dropped {col2}")
    else:
      print("    Invalid choice. No variable dropped.")

  print("\nDataFrame after dropping correlated variables:")
  print(df_encoded.head())

elif choice == "2":
  # Combine correlated variables into a new feature
  print("\nYou chose to combine correlated variables into a new feature.")
  print("This requires further analysis and feature engineering.")
  print("For now, we'll simply create a new feature by averaging the values of the correlated variables.")
  for col1, col2, correlation in highly_correlated_pairs:
    new_column_name = f"{col1}_{col2}_combined"
    df_encoded[new_column_name] = (df_encoded[col1] + df_encoded[col2]) / 2
    print(f"  - Created new feature: {new_column_name}")

  print("\nDataFrame after combining correlated variables:")
  print(df_encoded.head())

else:
  print("Invalid choice.")

# Allow user to undo changes
print("\nDo you want to undo the changes?")
undo_choice = input("Enter 'yes' to undo or any other key to keep the changes: ")
if undo_choice.lower() == "yes":
  # Reset the DataFrame to its state before handling multicollinearity
  # (This would involve reloading the data or restoring a previous state)
  print("Changes undone.")
else:
  print("Changes kept.")

# Summary
print("\nOverview Summary:")
print("Multicollinearity addressed.")
print("Method used:", "Dropping variables" if choice == "1" else "Combining variables")
print("Note: Dealing with multicollinearity is an iterative process.")
print("  - Consider the specific context and goals of your analysis.")
print("  - Evaluate the impact of different approaches on model performance.")
print("  - Explore advanced techniques like Principal Component Analysis (PCA) for dimensionality reduction.")
```
