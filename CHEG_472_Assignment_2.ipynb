```python
# prints '- no break' if loop completed without break

for i in range(10):
    if i == 12:
        break
    print (i, end=" ")
else:
    print ('- no break')
```

```python
# skips even numbers, but continues through loop after

for i in range(10):
    if i % 2 == 0:
        continue
    print (i, end=" ")
```

```python
# while loop

a = 0
while a < 10:
    print (a, end=" ")
    a += 1
```
